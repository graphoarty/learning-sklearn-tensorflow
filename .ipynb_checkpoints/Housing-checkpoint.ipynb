```python
import random
import matplotlib.pyplot as plt
import numpy as np
x = np.array([x for x in range(0, 10)])
y = np.array(random.sample(range(10000), 10))
x2 = np.array([x2 for x2 in range(0, 10)])
y2 = np.array(random.sample(range(10000), 10))
plt.plot(x, y, label='First Line')
plt.plot(x2, y2, label='Second Line')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Insanity Graph')
plt.legend()
plt.show()
```
